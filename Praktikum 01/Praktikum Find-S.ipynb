```python
import pandas as pd
import numpy as np

dataset = pd.read_csv('sport.csv')
print(dataset)
```

```python
dataset = pd.read_csv('sport.csv')
datatraining = np.array(dataset)[:, 1:-1]
print("Data training:\n", datatraining)
```

```python
target = np.array(dataset)[:, -1]
print("Target:\n", target)
```

```python
def train(c, t):

  # Melakukan iterasi melalui setiap elemen dalam 't' (target)
  # 'enumerate' juga memberikan indeks dari elemen saat ini
  for i, val in enumerate(t):
    # Jika elemen target saat ini adalah 'Yes'
    if val == 'Yes':
      # Membuat salinan dari baris data training yang sesuai
      # dan menyimpannya sebagai hipotesa awal
      hipotesa = c[i].copy()

  # Melakukan iterasi melalui setiap baris dalam 'c' (data training)
  for i, val in enumerate(c):
    # Jika target untuk baris data training saat ini adalah 'Yes'
    if t[i] == 'Yes':
      # Melakukan iterasi melalui setiap elemen dalam hipotesa
      for x in range(len(hipotesa)):
        # Jika elemen dalam baris data training saat ini tidak sama
        # dengan elemen yang sesuai dalam hipotesa
        if val[x] != hipotesa[x]:
          hipotesa[x] = '?'

  return hipotesa

# Memanggil fungsi 'train' dengan 'datatraining' dan 'target' sebagai argumen
# dan menyimpan hasilnya dalam variabel 'hipotesa'
hipotesa = train(datatraining, target)
print("Hipotesa:\n", hipotesa)
```

```python
datatest = ['Sunny', 'Warm', 'Normal', 'Strong', 'Cool', 'Same']
print("Datatest:\n", datatest)
```

```python
def predict(datatest, hipotesa):
  for i in range(len(datatest)):
    if datatest[i] != hipotesa[i] and hipotesa[i] != '?':
      return "Jangan berlolahraga"
    return "Silakan berolahraga"

hasilPrediksi = predict(datatest, hipotesa)
print("Datatest:\n", datatest)
print("\nPrediksi:\n", hasilPrediksi)
```

```python
datatest = ['Rainy', 'Warm', 'Normal', 'Strong', 'Cool', 'Change']

hasilPrediksi = predict(datatest, hipotesa)
print("Datatest:\n", datatest)
print("\nPrediksi:\n", hasilPrediksi)
```
